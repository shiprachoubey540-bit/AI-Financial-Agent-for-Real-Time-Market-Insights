```python
import pandas as pd
import re
```

```python
df = pd.read_csv("/content/youtube_videos_zomato.csv")
df.head()
```

```python
print(f"Shape: {df.shape}")
print(df.dtypes)
```

```python
df["published_at"] = pd.to_datetime(df["published_at"], utc=True)
```

```python
df["published_date"] = df["published_at"].dt.date
df["year"]           = df["published_at"].dt.year
df["month"]          = df["published_at"].dt.month
df["day_of_week"]    = df["published_at"].dt.day_name()
```

```python
str_cols = ["video_id", "title", "channel", "search_query"]
df[str_cols] = df[str_cols].apply(lambda col: col.str.strip())
```

```python
df["title_clean"] = (
    df["title"]
    .str.lower()
    .str.replace(r"[^\w\s]", " ", regex=True)
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
)

df["channel_clean"] = df["channel"].str.lower().str.strip()
```

```python
before = len(df)
df.drop_duplicates(subset="video_id", keep="first", inplace=True)
df.reset_index(drop=True, inplace=True)
print(f"Dropped {before - len(df)} duplicate video_id rows")
```

```python
df["title_length"]    = df["title"].str.len()
df["title_word_count"] = df["title"].str.split().str.len()
```

```python
df["search_query_cat"] = df["search_query"].astype("category")
df["search_query_id"]  = df["search_query_cat"].cat.codes
```

```python
print("\nNull counts after preprocessing:")
print(df.isnull().sum())

print("\nFinal shape:", df.shape)
print(df.head(3).to_string())
```
